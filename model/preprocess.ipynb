```python
import numpy as np
import librosa.display
import torch
import librosa
import padasip as pa
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
y,sr= librosa.load('..//dataset//AEC-Challenge//datasets//real//__BRZi2uvkGpxHu808RzOg_nearend_singletalk_mic.wav')
```

```python
dlen=64*64
x = 0.3*librosa.tone(440, sr=22050, length=dlen)+0.2*librosa.tone(330, sr=22050, length=dlen)+0.1*librosa.tone(550, sr=22050, length=dlen)
d= librosa.tone(440, sr=22050, length=dlen)
```

```python
x=x.reshape((int(dlen/64),64))
d=d.reshape((int(dlen/64),64))
```

```python
h = pa.filters.AdaptiveFilter(model="NLMS", n=64, mu=0.1, w="random")
```

```python
yhat=h.run(x,d)
```

```python
fig, ax = plt.subplots(nrows=3, sharex=True)
librosa.display.waveshow(x, sr=sr, ax=ax[0])
librosa.display.waveshow(d, sr=sr, ax=ax[1])
librosa.display.waveshow(yhat, sr=sr, ax=ax[2])
ax[0].set(title='signals')
ax[0].label_outer()
```
